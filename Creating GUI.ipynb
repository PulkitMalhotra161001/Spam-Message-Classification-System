### Load packages

```python
import pickle
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import os
from tkinter import *
import pandas as pd
import numpy as np 
from tkinter import filedialog
```

### Load saved model

```python
model = pickle.load(open("Models/model.pkl","rb"))
cv = pickle.load(open("Models/cv.pkl","rb"))
```

### Create Tkinter GUI for Sentiment Analysis

```python
class MailSystem:
    def __init__(self):
        self.win1=tk.Tk()
        self.win1.title("MESSAGE BOX")
        self.rw = self.win1.winfo_screenwidth()
        self.rh = self.win1.winfo_screenheight()
        self.win1.geometry("%dx%d+0+0"%(self.rw,self.rh))
        self.win1.config(bg="#C1CEFE")
        self.top_frame=tk.Frame(self.win1,height=100,bg="gray20",width=self.rw,bd=5,relief="groove")
        self.top_frame.pack(side="top",fill="both")
        self.lb = tk.Label(self.top_frame,text="Message Box",font=("times new roman",36,"bold"),fg="white",bg="#08605F")
        self.lb.pack(fill="both")        

        self.b1=tk.Button(self.win1,text="Login",width=15,font=("times new roman",20),fg="white",bg = "green",activebackground="gray40",pady=8,padx=14,command=self.login)
        self.b2=tk.Button(self.win1,text="Register",width=15,font=("times new roman",20),fg="white",bg = "green",activebackground="gray40",pady=8,padx=14,command=self.register)

        self.b1.place(x=530,y=250)
        self.b2.place(x=530,y=350)
        self.win1.mainloop()

    def login(self):
        self.win2=tk.Tk()
        self.win2.title("LOGIN")
        self.win2.geometry("500x400")
        self.win2.resizable(0,0)

        self.fr1 = tk.Frame(self.win2,height=80,width=400,bg="#08605F")
        self.fr1.pack(side="top",fill="both")
        self.lb1 = tk.Label(self.fr1,text="Login",font=("times new roman",18,"italic"),bg="#08605F",fg="white")
        self.lb1.pack()
        self.fr2 = tk.Frame(self.win2,width=400,height=600,bg="#C1CEFE")
        self.fr2.pack(fill="both")
        
        self.c1=tk.StringVar()
        self.c2=tk.StringVar()
        
        self.b11=tk.Label(self.win2,text="Contact Number: ",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        self.b12=tk.Label(self.win2,text="Password: ",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        
        self.b13=tk.Entry(self.win2,textvariable=self.c1,font=("times new roman",15,"bold"))
        self.b14=tk.Entry(self.win2,show="*",textvariable=self.c2,font=("times new roman",15,"bold")) 
        
        self.b15=tk.Button(self.win2,text="Login",width=10,font=("times new roman",16,"bold"),bg="green",fg="white",activebackground="gray85",command=self.profile)
        
        self.b11.place(x=40,y=100)
        self.b12.place(x=40,y=160)
        
        self.b13.place(x=220,y=100)
        self.b14.place(x=220,y=160)
        
        self.b15.place(x=170,y=280)
        
        self.win2.mainloop()
       
    def profile(self):
        v1=self.b13.get()
        v2=self.b14.get()
        #self.b13.delete(0,END)
        self.b14.delete(0,END)
        
        if v1 in os.listdir("Users/"):
            cfile="Users/{}/Profiles.csv".format(v1)
            self.data = pd.read_csv(cfile)
            if str(self.data.Password.sum())==str(v2):
                self.win3=tk.Tk()
                self.win3.title("PROFILE")
                self.win3.geometry("500x500")
                self.win3.config(bg="#C1CEFE")
                self.win3.resizable(0,0)

                self.b111=tk.Button(self.win3,text="View Profile",bg="#04A777",fg = "white",width=15,font=("times new roman",14),command=self.vprofile)
                self.b112=tk.Button(self.win3,text="Send Message",bg="#04A777",fg = "white",width=15,font=("times new roman",14),command=self.sms)
                self.b113=tk.Button(self.win3,text="Inbox",bg="#04A777",fg = "white",width=15,font=("times new roman",14),command=self.inbox)
                self.b114=tk.Button(self.win3,text="Logout",bg="#A31621",fg = "white",width=15,font=("times new roman",14),command=self.logout)
                self.b115=tk.Button(self.win3,text="Spam",bg="#04A777",fg = "white",width=15,font=("times new roman",14),command=self.spam)

                self.b111.place(x=180,y=90)
                self.b112.place(x=180,y=160)
                self.b113.place(x=180,y=230)
                self.b115.place(x=180,y=300)
                self.b114.place(x=180,y=370)
                self.win3.mainloop()
                
            else:
                messagebox.showerror("Error","Incorrect Password, Please Try Again")
        else:
            messagebox.showerror("User Not Found","User with This Contact Number is not in our Databse")
   
    def vprofile(self):
        self.win5=tk.Tk()
        self.win5.title("USER PROFILE")
        self.win5.geometry("500x500")
        self.win5.config(bg="#C1CEFE")
        self.win5.resizable(0,0)
        
        self.fr15 = tk.Frame(self.win5,height=80,width=400,bg="#08605F")
        self.fr15.pack(side="top",fill="both")        
        
        self.b1110=tk.Label(self.fr15,text="USER PROFILE",bg="#08605F",fg="white",font=("times new roman",18,"bold"))
        
        self.b1111=tk.Label(self.win5,text="NAME:",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        self.b1112=tk.Label(self.win5,text="CONTACT NO.:",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        self.b1113=tk.Label(self.win5,text="EMAIL ID:",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        self.b1114=tk.Label(self.win5,text="PASSWORD:",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        
        self.b1115=tk.Label(self.win5,text=self.data.Name.sum(),bg="#C1CEFE",fg="black",font=("times new roman",15))
        self.b1116=tk.Label(self.win5,text=self.data.Contact.sum(),bg="#C1CEFE",fg="black",font=("times new roman",15))
        self.b1117=tk.Label(self.win5,text=self.data.Mail.sum(),bg="#C1CEFE",fg="black",font=("times new roman",15))
        self.b1118=tk.Label(self.win5,text=self.data.Password.sum(),bg="#C1CEFE",fg="black",font=("times new roman",15))

        self.b1110.pack()
        
        self.b1111.place(x=40,y=100)
        self.b1112.place(x=40,y=150)
        self.b1113.place(x=40,y=200)
        self.b1114.place(x=40,y=250)
        
        self.b1115.place(x=230,y=100)
        self.b1116.place(x=230,y=150)
        self.b1117.place(x=230,y=200)
        self.b1118.place(x=230,y=250)
        
        self.win5.mainloop()

    def sms(self):
        self.win6=tk.Tk()
        self.win6.title("COMPOSE")
        self.win6.geometry("500x500")
        self.win6.config(bg="#C1CEFE")
        self.win6.resizable(0,0)
        
        self.c7=tk.StringVar()
        #self.c8=tk.StringVar()
        self.c9=tk.StringVar()
        
        self.b1121=tk.Label(self.win6,text="CONTACT NO. :",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        #self.b1122=tk.Label(self.win6,text="EMAIL ID:",bg="gray85",fg="black",font=("times new roman",15,"bold"))
        self.b1123=tk.Label(self.win6,text="SUBJECT:",bg="#C1CEFE",fg="black",font=("times new roman",15,"bold"))
        
        self.b1124=tk.Entry(self.win6,textvariable=self.c7,font=("times new roman",15,"bold"))
        #self.b1125=tk.Entry(self.win6,textvariable=self.c8,font=("times new roman",15,"bold"))
        self.b1126=tk.Entry(self.win6,textvariable=self.c9,font=("times new roman",15,"bold"))
        
        self.b1127 = tk.Text(self.win6,font=("times new roman",15),height=9,width=47)
        
        self.b1128=tk.Button(self.win6,text="Send",bg="green",activebackground="gray40",fg="white",width=15,font=("times new roman",16,"bold"),command=self.sendsms)
        
        self.b1121.place(x=10,y=30)
        #self.b1122.place(x=10,y=50)
        self.b1123.place(x=10,y=80)

        self.b1124.place(x=200,y=30)
        #self.b1125.place(x=200,y=50)
        self.b1126.place(x=200,y=80)
        self.b1127.place(x=10,y=130)
        self.b1128.place(x=180,y=400)
        self.win6.mainloop()
        
    def sendsms(self):
        v7=self.b1124.get()
        #v8=self.b1125.get()
        v9=self.b1126.get()
        v10=self.b1127.get("1.0",END)
        
        if v7 in os.listdir("Users/"):
            dfile="Users/{}/Profiles.csv".format(v7)
            self.data1 = pd.read_csv(dfile)
            #if self.data1.Mail.sum()==v8:
            
            if "Spam" not in os.listdir("Users/{}/".format(v7)):
                os.mkdir("Users/{}/Spam".format(v7))
            if "Ham" not in os.listdir("Users/{}/".format(v7)):
                os.mkdir("Users/{}/Ham".format(v7))

            self.converted = cv.transform([v10])
            self.pred = model.predict(self.converted)[0]
            self.Label = {0:"HAM", 1:"SPAM"}
            
            if self.Label[self.pred]=="HAM":            
                    p1="Users/{}/Ham/{}.txt".format(v7,v9)
                    f1=open(p1,'x')
                    str2=""
                    str2+="{}".format(v10)
                    f1.write(str2)
                    f1.close()
            
            if self.Label[self.pred]=="SPAM":
                p2="Users/{}/Spam/{}.txt".format(v7,v9)
                f2=open(p2,'x')
                str3=""
                str3+="{}".format(v10)
                f2.write(str3)
                f2.close()
                self.win6.destroy()
            messagebox.showinfo("Done","Message Send") 
            
        else:
            messagebox.showerror("User Not Found","User with This Contact Number is not in our Database")
        
    
    def inbox(self):
        self.win7=tk.Tk()
        self.win7.title("INBOX")
        self.win7.geometry("500x500")
        self.win7.config(bg="#C1CEFE")
        self.win7.resizable(0,0)
        self.label0=tk.Button(self.win7,text="Inbox Messages",fg="white",bg ="green",activebackground="gray40",font=("times new roman",14),command=self.ham)
        self.label0.place(x=10,y=50)
        self.win7.mainloop()
        
    def ham(self):
        v11=self.b13.get()
        a="Users/{}/Ham/".format(v11)
        file_name = filedialog.askopenfilename(initialdir=a, title="Select A File", filetype=(("txt","*.txt"),("jpg","*.jpg")))
        self.label1 = tk.Label(self.win7, text="")
        self.label1.place(x=500,y=500)
        self.label2=tk.Text(self.win7,height=15,width=60)
        path="{}".format(file_name)
        f=open(path,'r')
        lines = f.readlines()
        con=""
        for line in lines:
            con+=line
        self.label2.insert(tk.END, con)
        self.label2.place(x=10,y=100)
        self.label2.configure(state="disabled")
        
    
    def spam(self):
        self.win8=tk.Tk()
        self.win8.title("SPAM")
        self.win8.geometry("500x500")
        self.win8.config(bg="#C1CEFE")
        self.win8.resizable(0,0)
        self.label3=tk.Button(self.win8,text="Spam Folder",fg="white",bg ="green",activebackground="gray40",font=("times new roman",14),command=self.sppam)
        self.label3.place(x=10,y=50)
        self.win8.mainloop()
        
    def sppam(self):
        v12=self.b13.get()
        b="Users/{}/Spam/".format(v12)
        file_namee = filedialog.askopenfilename(initialdir=b, title="Select A File", filetype=(("txt","*.txt"),("jpg","*.jpg")))
        self.label4 = tk.Label(self.win8, text="")
        self.label4.place(x=500,y=500)
        self.label5=tk.Text(self.win8,height=15,width=60)
        pathh="{}".format(file_namee)
        ff=open(pathh,'r')
        liness = ff.readlines()
        conn=""
        for line in liness:
            conn+=line
        self.label5.insert(tk.END, conn)
        self.label5.place(x=10,y=100)
        self.label5.configure(state="disabled")

    def logout(self):
        self.win3.destroy()
        
    def register(self):
        self.win4=tk.Tk()
        self.win4.title("REGISTER")
        self.win4.geometry("400x550")
        self.win4.resizable(0,0)
        self.tp_fr = tk.Frame(self.win4,height=80,width=400,bg="#08605F")
        self.tp_fr.pack(side="top",fill="both")
        self.lb = tk.Label(self.tp_fr,text="Register",font=("times new roman",18,"italic"),bg="#08605F",fg="white")
        self.lb.pack()
        self.bt_fr = tk.Frame(self.win4,width=400,height=600,bg="#C1CEFE")
        self.bt_fr.pack(fill="both")

        self.c3=tk.StringVar()
        self.c4=tk.StringVar()
        self.c5=tk.StringVar()
        self.c6=tk.StringVar()

        self.b21=tk.Label(self.win4,text="Name: ",bg="#C1CEFE",fg="black",font=("times new roman",14,"bold"))
        self.b22=tk.Label(self.win4,text="Contact Number: ",bg="#C1CEFE",fg="black",font=("times new roman",14,"bold"))
        self.b23=tk.Label(self.win4,text="E-mail: ",bg="#C1CEFE",fg="black",font=("times new roman",14,"bold"))
        self.b24=tk.Label(self.win4,text="Password: ",bg="#C1CEFE",fg="black",font=("times new roman",14,"bold"))
        
        self.b25=tk.Entry(self.win4,textvariable=self.c3,font=("times new roman",14))
        self.b26=tk.Entry(self.win4,textvariable=self.c4,font=("times new roman",14))
        self.b27=tk.Entry(self.win4,textvariable=self.c5,font=("times new roman",14))
        self.b28=tk.Entry(self.win4,show="*",textvariable=self.c6,font=("times new roman",14))

        self.b29=tk.Button(self.win4,text="Register",font=("times new roman",20,"bold"),fg="white",bg="green",activebackground="SpringGreen2",command=self.registration)
        self.b29.place(x=140,y=350)

        self.b21.place(x=20,y=80)
        self.b22.place(x=20,y=140)
        self.b23.place(x=20,y=200)
        self.b24.place(x=20,y=260)
        
        self.b25.place(x=180,y=80)
        self.b26.place(x=180,y=140)
        self.b27.place(x=180,y=200)
        self.b28.place(x=180,y=260)
        
        self.win4.mainloop()
    
    def registration(self):
        v3=self.b25.get()
        v4=self.b26.get()
        v5=self.b27.get()
        v6=self.b28.get()
        if v4.isdigit()==True:
            if len(v4)==10:
                if "@gmail.com" not in v5:
                    messagebox.showerror("Mail Address Error","Not a valid Email address")
                else:
                    if "Users" not in os.listdir():
                        os.mkdir("Users")
                    if v4 not in os.listdir("Users/"):
                        os.mkdir("Users/{}".format(v4))
                    else:
                        messagebox.showerror("Format Error","A User with this contact number already exists!")

                    path="Users/{}/Profiles.csv".format(v4)
                    f=open(path,'x')
                    st1=""
                    st1+="Name,Contact,Mail,Password"
                    st1+="\n"
                    st1+="{},".format(v3)
                    st1+="{},".format(v4)
                    st1+="{},".format(v5)
                    st1+="{}".format(v6)
                    f.write(st1)
                    f.close()
                    self.win4.destroy()
                    messagebox.showinfo("Success","Data Registered Successfully")
            else:
                messagebox.showerror("Phone No. Error","Not a valid Contact Number")
        else:
            messagebox.showerror("Value Error","Contact Number can only be in digits")
            
obj = MailSystem()
```
